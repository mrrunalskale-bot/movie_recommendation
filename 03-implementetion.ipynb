```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
ds = pd.read_csv('clean_data.csv')
```

```python
ds.info()
```

```python
ds = ds.drop(columns=['Unnamed: 0'])
```

```python
ds.info()
```

```python
ds.columns
```

```python
x = ds[['name', 'company', 'year', 'kms_driven', 'fuel_type']]
```

```python
y = ds[['Price']]
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit(ds[['name', 'company','fuel_type']])
```

```python
ohe.categories_
```

```python
 ct = make_column_transformer((OneHotEncoder(categories=ohe.categories_),["name","company","fuel_type"]),remainder="passthrough")
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(ct,reg)
```

```python
pipe
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.10,random_state=0)
```

```python
pipe.fit(x_train,y_train)
```

```python
input_data = pd.DataFrame(data=[["Grand i10 Magna 1.2 Kappa VTVT","Hyundai",2014,28000.0,"Petrol"]],
                     columns=['name', 'company', 'year', 'kms_driven', 'fuel_type'])
```

```python
pipe.predict(input_data)
```

```python
ds.info()
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred = pd.DataFrame(data=y_pred,columns=['Result'])
```

```python
result = pd.concat([y_pred.reset_index(drop=True),y_test.reset_index(drop=True)],axis=1)
result
```

```python
from sklearn.metrics import r2_score
score = r2_score(result['Price'],result['Result'])
score
```

```python
scores = []
for i in range(0,101):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=i)

    pipe.fit(x_train,y_train)
    y_pred = pipe.predict(x_test)
    y_pred = pd.DataFrame(data=y_pred,columns=['result'])
    result = pd.concat([y_test.reset_index(drop= True),y_pred.reset_index(drop= True)],axis=1)

    score = r2_score(result['Price'],result['result'])
    scores.append(score)
```

```python
max(scores)


index = np.argmax(scores)
print(index)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=90)
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
y_pred = pd.DataFrame(data=y_pred,columns=['result'])
result = pd.concat([y_test.reset_index(drop= True),y_pred.reset_index(drop= True)],axis=1)
score = r2_score(result['Price'],result['result'])
score*100
```
